```python
import pandas as pd

# Read the file into a DataFrame
df = pd.read_csv('game_log.csv')
total_games = len(df)
wins = df['winner'].value_counts()
draws = df[df['winner'] == 'draw']
num_draws = len(draws)

# Displaying the statistics in a table
stats_df = pd.DataFrame({
    'Statistic': ['Total Games', 'X Wins', 'O Wins', 'Draws'],
    'Value': [total_games, wins.get('X', 0), wins.get('O', 0), num_draws]
})
print(stats_df)
```

```python
import matplotlib.pyplot as plt

# Bar chart for wins
wins.plot(kind='bar')
plt.title('Number of Wins for Each Player')
plt.xlabel('Player')
plt.ylabel('Wins')
plt.show()

# Pie chart for game outcomes
outcome_counts = df['winner'].value_counts()
outcome_counts.plot(kind='pie', autopct='%1.1f%%')
plt.title('Game Outcomes')
plt.show()

# Line graph for move number
df['move_number'].plot(kind='hist', bins=range(1, df['move_number'].max()+2), edgecolor='black')
plt.title('Distribution of Game Lengths by Move Number')
plt.xlabel('Move Number')
plt.ylabel('Number of Games')
plt.xticks(range(1, df['move_number'].max()+1))
plt.show()
```
